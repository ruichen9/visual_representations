```python
import pandas as pd
import numpy as np
```

```python
rate = pd.read_csv("Survey_raw.csv")
```

```python
rate.head()
```

```python
rate = rate.iloc[:, range(17, 53)]

clean_rate = rate.iloc[3:36, :]

rate = clean_rate.to_numpy()

rate = rate.astype(np.float)

par_avg = np.nanmean(rate, axis = 0)

par_avg.shape
```

```python
np.savetxt("avg.csv", par_avg, delimiter=",")
```
