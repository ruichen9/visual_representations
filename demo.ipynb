```python
model_list = {}

model_list["alexnet"] = models.alexnet(pretrained=True)


model_list["vgg11"] = models.vgg11(pretrained=True)
model_list["vgg11_bn"] = models.vgg11_bn(pretrained=True)
model_list["vgg13"] = models.vgg13(pretrained=True)
model_list["vgg13_bn"] = models.vgg13_bn(pretrained=True)
model_list["vgg16"] = models.vgg16(pretrained=True)
model_list["vgg16_bn"] = models.vgg16_bn(pretrained=True)
model_list["vgg19"] = models.vgg19(pretrained=True)
model_list["vgg19_bn"] = models.vgg19_bn(pretrained=True)

model_list["resnet18"] = models.resnet18(pretrained=True)
model_list["resnet34"] = models.resnet34(pretrained=True)
model_list["resnet50"] = models.resnet50(pretrained=True)
model_list["resnet101"] = models.resnet101(pretrained=True)
model_list["resnet152"] = models.resnet152(pretrained=True)

model_list["vit_b_16"] = models.vit_b_16(pretrained=True)
model_list["vit_b_32"] = models.vit_b_32(pretrained=True)
model_list["vit_l_16"] = models.vit_l_16(pretrained=True)
model_list["vit_l_32"] = models.vit_l_32(pretrained=True)

model_list["efficientnet_b0"] = models.efficientnet_b0(pretrained=True)
model_list["efficientnet_b1"] = models.efficientnet_b1(pretrained=True)
model_list["efficientnet_b2"] = models.efficientnet_b2(pretrained=True)
model_list["efficientnet_b3"] = models.efficientnet_b3(pretrained=True)
model_list["efficientnet_b4"] = models.efficientnet_b4(pretrained=True)
model_list["efficientnet_b5"] = models.efficientnet_b5(pretrained=True)
model_list["efficientnet_b6"] = models.efficientnet_b6(pretrained=True)
model_list["efficientnet_b7"] = models.efficientnet_b7(pretrained=True)
```

```python
def get_results(file_holder, class_idx, type = 'prob'):
    files_list = glob_file(file_holder,'.png')
    data_list= get_data(files_list, pic_size=224)
    results = {}
    for model_name, model in model_list.items():
        outputs = get_outputs(data_list, model, type = type)
        scores = get_class_score(outputs, class_idx=class_idx)
        results[model_name] = scores
    return results
```

```python
# class name - index
# desk 526
# seastar 327
# guinea pig 338

desk_y = get_results('data/Desk', 526, type='raw')
desk_z = get_results('data/Desk', 526, type='prob')
sea_star_y = get_results('data/Sea_Star', 327, type='raw')
sea_star_z = get_results('data/Sea_Star', 327, type='prob')
guinea_pig_y = get_results('data/Guinea_Pig', 338, type='raw')
guinea_pig_z = get_results('data/Guinea_Pig', 338, type='prob')
```

```python
import pickle

def namestr(obj, namespace):
    return [name for name in namespace if namespace[name] is obj]

def write_result(dict):
    f = open(namestr(dict,globals())[0]+".pkl","wb")
    pickle.dump(dict,f)
    f.close()
```

```python
write_result(desk_y)
write_result(desk_z)
write_result(sea_star_y)
write_result(sea_star_z)
write_result(guinea_pig_y)
write_result(guinea_pig_z)
```
